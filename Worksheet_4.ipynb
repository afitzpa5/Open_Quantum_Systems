4. Now let us construct a Liouvillian numerically. Consider the Lindblad equation for a qubit in a thermal bath $$ \frac{{\rm d}}{{\rm d}t} \hat{\rho} = \frac{1}{{\rm i}\hbar} [ \hat{H},\hat{\rho}] + \gamma (1+\bar{n})\mathcal{D}[\hat{\sigma}_-]\hat{\rho} + \gamma \bar{n} \mathcal{D}[\hat{\sigma}_+]\hat{\rho} = \mathcal{L}\hat{\rho},$$ where $\hat{H} = \hbar\omega_0 \hat{\sigma}_z/2$, $\bar{n} = ({\rm e}^{\beta\hbar\omega_0}-1)^{-1}$ and $\beta = 1/k_BT$ is the inverse temperature. Take the parameters to be $\omega_0=1$, $\gamma = 0.1\omega_0$ and $k_BT = 0.5\hbar\omega_0$.

(a) The Pauli operator basis is defined by $\hat{e}_j = \hat{\sigma}_j/\sqrt{2}$, where $\{\hat{\sigma}_0,\hat{\sigma}_1,\hat{\sigma}_2,\hat{\sigma}_3\}= \{\hat{\mathbb{1}},\hat{\sigma}_x,\hat{\sigma}_y,\hat{\sigma}_z\}$. Verify numerically that this forms an orthonormal basis with respect to the inner product $\langle \hat{e}_i,\hat{e}_j\rangle = {\rm tr}(\hat{e}_i^\dagger \hat{e}_j)$.

```python
#first define the identity and pauli matrices
iden, x, y, z = np.eye(2), np.array([[0,1],[1,0]]), 1j*np.array([[0,-1],[1,0]]),np.array([[1,0],[0,-1]]) 

#define the pauli basis
pauli = 1/(np.sqrt(2))*np.array([iden,x,y,z])

#if I take the inner product of all the combinations and print these should equal delta_ij
#i'll define a function to do the inner product between to elements

def inner(x,y):
    xt = x.T.conj()
    
    if np.ndim(xt@y)>=2:
        return(np.trace(xt@y))
    else:                       #function works for vectors and matrices
        return xt@y

#nelow should print 1 for inner product of basis vector with itself and 0 for all others
for i in range(4):
    for j in range(4):
        print(f'<e_{i},e_{j}> = ', inner(pauli[i],pauli[j]))
print('Results are as expected')
```

(b) Construct and print the Liouvillian operator $\hat{\mathcal{L}}$ as a NumPy array in the Pauli basis.

```python
#first I will define functions to return dissipators and commutators

def com(x,y):
    return x@y-y@x

def diss(x,y):
    return x@y@x.T.conj()-0.5*(x.T.conj()@x@y+y@x.T.conj()@x)    

#I will also define the pauli +- operators
plus,minus = 1/2*(x+1j*y),1/2*(x-1j*y)

#next I define the relevent parameters
w0,gamma = 1,0.1
beta = 1/(0.5*w0)
H = w0*z/2
nbar = (np.exp(beta)-1)**-1
print(H)
```

Now we note that the matrix elements of the Liouvillian are given by 
$$\mathcal{L_{ij}} = \langle \hat e_{i},\mathcal{L(\hat e_{j}) \rangle}$$
which we will use to generate the matrix below

```python
L = np.zeros((4,4),dtype='complex')

for i in range(4):
    for j in range(4):
        l=-1j*com(H,pauli[j])+gamma*(1+nbar)*diss(minus,pauli[j])+gamma*nbar*diss(plus,pauli[j])
        L[i][j]+=(inner(pauli[i],l))

print(L)
```


(c) Find the eigenvalues and right eigenvectors of $\hat{\mathcal{L}}$. Print the numerically obtained eigenvectors in their standard matrix form (i.e. expressed in the computational basis $|0\rangle$ and $|1\rangle$). Verify that the steady state is thermal as expected. Comment on the implications of these results for the time evolution of the qubit.

```python
#the steady state corresponds to the eigenvector whos eigenvalue is 0, so we print this
eig1 = scipy.linalg.eig(L,left=False,right=True)
vec = eig1[1][:,np.where(abs(eig1[0])<0.00001)[0][0]].flatten()

#now this eigenvector is in the pauli basis, so we have to convert to computational basis

def to_comp(vector,basis):
    return np.array([vector@basis1.reshape(4,1) for basis1 in basis]).flatten()

#now can write the eigenvector in the computational basis, though it needs to be properly normalised
vec = (to_comp(vec,pauli)/np.trace(to_comp(vec,pauli).reshape(2,2))).reshape(2,2)

print('Steady state in computational basis\n',vec)

print('Thermal state for beta = 2 \n',expm(-beta*H)/np.trace(expm(-beta*H)))

print('Can see the steady state is thermal as expected')
```

As $t\rightarrow \infty$ we expec the qubit to evolve to a thermal state with inverse temperature of $\beta = 2$. This makes sense as due to this being the 0 eigenvalue state, it means the rate time evolution of the density matrix is now 0 i.e unchanging. $$ |\tilde{\psi}(t+\Delta t)\rangle = {\rm e}^{-{\rm i}\hat{H}_{\rm eff}\Delta t /\hbar}|\tilde{\psi}(t)\rangle.$$
Plot a histogram of the numerically obtained $W(t)$ as a function of $\gamma t$ for 200 trajectories, and quantitatively compare it with your analytical result. 

```python
#first some initialisations, we note H = 0 in the above case
gamma = 0.2
t, dt = 150, 0.05
Heff = -1j*gamma/2*(plus@minus)

#define some evolution operator 
def evo(H,dt):
    return expm(-1j*H*dt)

#now we need a list of times, if we set gamma*t = 30 we have t = 150
timesteps = np.linspace(0,t,int(t/dt))
xaxis = gamma*timesteps


#now create initial state
zero = np.array([1,0])
#now the list of evolved states
states = np.array([evo(Heff,t1)@zero for t1 in timesteps])

#now calculate W for each of these states, via a simple finite difference 
innerprod = np.array([np.real(inner(state,state)) for state in states])
waiting = -np.array([(innerprod[i+1]-innerprod[i])/(2*dt) for i in range(len(innerprod)-1)])


    
#analytical expression is given by
plt.plot(xaxis,gamma*np.exp(-gamma*timesteps),label='analytical',marker='o',markersize=3)
plt.plot(xaxis[:-1],waiting,label='numerical',color='r')
plt.title('Analytical vs Numerical Waiting Time Distribution')
plt.xlabel('$\gamma t$')
plt.ylabel('W(t)')
plt.legend()
plt.show()
```

Consider the anharmonic oscillator with Hamiltonian given in the interaction picture by $$ \hat{H} = \hbar\Delta \hat{a}^\dagger\hat{a} + \hbar U\hat{a}^\dagger \hat{a}^\dagger \hat{a}\hat{a} + \frac{\hbar F}{2} (\hat{a} + \hat{a}^\dagger),$$ where $\Delta$ is the detuning of the drive from resonance, $U$ quantifies the anharmonicity and $F$ is the drive strength (the rotating wave approximation has been invoked). The system undergoes dissipation described by the Lindblad equation $$\frac{\rm d}{{\rm d}t} \hat{\rho} = \frac{1}{{\rm i}\hbar} [\hat{H},\hat{\rho}] + \gamma\mathcal{D}[\hat{a}]\hat{\rho}.$$ Find the steady state numerically and make a contour plot of its Husimi Q-function for $\Delta =0$ and $\gamma = 0.3F$, and with three values of anharmonicity: $U =0$, $U=0.1F$ and $U=F$. Comment briefly on the nature of the steady state in each case.

```python
#first we pick a dimention to work with a truncated hilbert space, and define our creation and annihilation operators
dim = 30
F, gamma, U , delta = 1, 0.3, [0,0.1,1], 0

def a(d):
    b = np.eye(d)
    m = np.zeros((d,d))
    for i in range(1,d):                          #creates annihilation operator in d dimenions
        m+=np.sqrt(i)*np.outer(b[i-1],b[i])
        
    return m

def ad(d):
    return a(d).T.conj()                         #conjugate transpose of a

#am going to take F=1 for simplicity and define the Hamiltonian
def Ham(d,delta,U,F):
    return delta*(ad(d)@a(d))+U*(ad(d)@ad(d)@a(d)@a(d))+F/2*(a(d)+ad(d))

H = [Ham(dim,delta,Us,F) for Us in U]

superH = [superop(H1,'left')-superop(H1,'right') for H1 in H]

def Liouvillian(h,gam,d):
    return -1j*h+gam*dissm1(a(d))

L = [Liouvillian(superH1,gamma,dim) for superH1 in superH]

eigsys = [np.linalg.eig(L1) for L1 in L]
ind = [np.where(abs(eigsys1[0])<0.00001) for eigsys1 in eigsys]  #find the indices of the 0 eigenvalue term

steady = []
for i in range(3):
    steady.append(eigsys[i][1][:,ind[i]].flatten().reshape(dim,dim))
    steady[i] = steady[i]/np.trace(steady[i])

#now define truncated displacement operator to create coherent states
def disp(d,alpha):
    return expm(alpha*ad(d)-np.conj(alpha)*a(d))


def Qs(alpha,rho,d):     #calculates the q function for a given complex number and state
    zero = np.zeros(d)
    zero[0]+=1
    ket = disp(d,alpha)@zero
    bra = ket.T.conj()
    
    return 1/(np.pi)*bra@rho@ket

def Q(alpha,rho,d):
    res = np.zeros((100,100))
    for i, x in enumerate(alpha):      #returns the q function over a range of complex numbers
        for j , y in enumerate(x):
            res[i][j] = np.real(Qs(y,rho,d))
    return res

#now define a range of complex numbers
alpha1 = np.linspace(-5,5,100)
X, Y = np.meshgrid(alpha1,alpha1)
compnums = X+1j*Y

norm = plt.Normalize(Q(compnums,steady[0],dim).min(),Q(compnums,steady[0],dim).max())  #create standard scale so the differences
                                                                                       #in plots are more easily seen
xlim = alpha1.min(),alpha1.max()
ylim = xlim
fig, (ax0, ax1, ax2) = plt.subplots(ncols=3, sharey=True, figsize=(15, 4))
fig.suptitle('Husimi Q functions for steady states')

hb = ax0.contourf(X,Y,Q(compnums,steady[0],dim),cmap='inferno', norm = norm) 
ax0.set(xlim=xlim, ylim=ylim)
ax0.set_title("U=0")
cb = fig.colorbar(hb, ax=ax0)

hb = ax1.contourf(X,Y,Q(compnums,steady[1],dim),cmap='inferno', norm = norm) 
ax1.set(xlim=xlim, ylim=ylim)
ax1.set_title("U=0.1")
cb = fig.colorbar(hb, ax=ax1)

hb = ax2.contourf(X,Y,Q(compnums,steady[2],dim),cmap='inferno', norm = norm) 
ax2.set(xlim=xlim, ylim=ylim)
ax2.set_title("U=1")
cb = fig.colorbar(hb, ax=ax2)
```

We can see as the anharmonicity increases, so to does the spread of the Q function distributions. Since the spreads overall are not that large, and given in the U=0 case the distribution is not centered on 0, it seems the steady state of this system is not thermal. It has the look of a coherent state with $$\alpha \approx 0-3.3i$$ which I plot below for comparison.

```python
def fock(n,d):
    f = np.zeros(d)       #returns the fock state |n> for a given n in dimension d 
    f[n]+=1
    return f

coherent = disp(dim,0-3.3j)@fock(0,dim)
coherent = np.outer(coherent,coherent.T.conj())

fig, (ax0, ax1) = plt.subplots(ncols=2, sharey=True, figsize=(12, 4))
fig.suptitle('Husimi Q functions for steady states')

hb = ax0.contourf(X,Y,Q(compnums,steady[0],dim),cmap='inferno') 
ax0.set(xlim=xlim, ylim=ylim)
ax0.set_title("U=0")
cb = fig.colorbar(hb, ax=ax0)

hb = ax1.contourf(X,Y,Q(compnums,coherent,dim),cmap='inferno') 
ax1.set(xlim=xlim, ylim=ylim)
ax1.set_title(r"Coherent state with $\alpha = 0-3.3i$")
cb = fig.colorbar(hb, ax=ax1)

plt.show()
```
